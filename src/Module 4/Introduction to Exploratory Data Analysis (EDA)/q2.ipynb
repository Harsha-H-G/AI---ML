**Title**: Handling Outliers<br>

**Task**: Detect and propose handling methods for outliers in the numerical columns of the datasets.<br>

Task 1: Age in Employee Dataset<br>
Task 2: Price in Product Sales Dataset<br>
Task 3: Math Score in Student Grades Dataset<br>

Instructions:<br>

Use box plots to visualize potential outliers.<br>
Suggest methods to handle them, such as removal or transformation.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Task 1: Age in Employee Dataset
age_data = pd.DataFrame({'Age': [22, 25, 28, 30, 32, 34, 36, 60, 100]})  # 100 is a potential outlier

plt.figure(figsize=(6, 2))
plt.boxplot(age_data['Age'], vert=False)
plt.title('Box Plot of Age (Employee Dataset)')
plt.xlabel('Age')
plt.show()

# Detect outliers using IQR method
Q1 = age_data['Age'].quantile(0.25)
Q3 = age_data['Age'].quantile(0.75)
IQR = Q3 - Q1
age_outliers = age_data[(age_data['Age'] < (Q1 - 1.5 * IQR)) | (age_data['Age'] > (Q3 + 1.5 * IQR))]
print("Age outliers:\n", age_outliers)

# Task 2: Price in Product Sales Dataset
price_data = pd.DataFrame({'Price': [10, 12, 15, 14, 13, 16, 18, 100]})  # 100 is a potential outlier

plt.figure(figsize=(6, 2))
plt.boxplot(price_data['Price'], vert=False)
plt.title('Box Plot of Price (Product Sales Dataset)')
plt.xlabel('Price')
plt.show()

Q1 = price_data['Price'].quantile(0.25)
Q3 = price_data['Price'].quantile(0.75)
IQR = Q3 - Q1
price_outliers = price_data[(price_data['Price'] < (Q1 - 1.5 * IQR)) | (price_data['Price'] > (Q3 + 1.5 * IQR))]
print("Price outliers:\n", price_outliers)

# Task 3: Math Score in Student Grades Dataset
math_score_data = pd.DataFrame({'Math Score': [55, 60, 65, 70, 75, 80, 85, 90, 100, 200]})  # 200 is a potential outlier

plt.figure(figsize=(6, 2))
plt.boxplot(math_score_data['Math Score'], vert=False)
plt.title('Box Plot of Math Score (Student Grades Dataset)')
plt.xlabel('Math Score')
plt.show()

Q1 = math_score_data['Math Score'].quantile(0.25)
Q3 = math_score_data['Math Score'].quantile(0.75)
IQR = Q3 - Q1
math_outliers = math_score_data[(math_score_data['Math Score'] < (Q1 - 1.5 * IQR)) | (math_score_data['Math Score'] > (Q3 + 1.5 * IQR))]
print("Math Score outliers:\n", math_outliers)

# Handling methods:
# - Remove outliers: Drop rows with outlier values.
# - Transform data: Apply log or square root transformation to reduce the effect of outliers.
# - Cap values: Replace extreme values with upper/lower percentile values (winsorization).
```

**Title**: Visualizing Data Distributions<br>

**Task**: Create visualizations for data distributions.<br>

Task 1: Histogram for Age in Employee Dataset<br>
Task 2: Distribution plot for Price in Product Sales Dataset<br>
Task 3: Histogram for Math Score in Student Grades Dataset

Instructions:<br>

Use matplotlib or seaborn in Python to create the plots.<br>
Comment on the skewness or normality of the distributions.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Task 1: Histogram for Age in Employee Dataset
age_data = pd.DataFrame({'Age': [22, 25, 28, 30, 32, 34, 36, 60, 100]})
plt.figure(figsize=(6, 3))
sns.histplot(age_data['Age'], bins=8, kde=True, color='skyblue')
plt.title('Histogram of Age (Employee Dataset)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
# Comment: The distribution is right-skewed due to the outlier at 100.

# Task 2: Distribution plot for Price in Product Sales Dataset
price_data = pd.DataFrame({'Price': [10, 12, 15, 14, 13, 16, 18, 100]})
plt.figure(figsize=(6, 3))
sns.histplot(price_data['Price'], bins=8, kde=True, color='orange')
plt.title('Distribution of Price (Product Sales Dataset)')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
# Comment: The distribution is highly right-skewed due to the outlier at 100.

# Task 3: Histogram for Math Score in Student Grades Dataset
math_score_data = pd.DataFrame({'Math Score': [55, 60, 65, 70, 75, 80, 85, 90, 100, 200]})
plt.figure(figsize=(6, 3))
sns.histplot(math_score_data['Math Score'], bins=10, kde=True, color='green')
plt.title('Histogram of Math Score (Student Grades Dataset)')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
# Comment: The distribution is right-skewed due to the outlier at 200.
```

**Title**: Finding Relationships Between Features<br>

**Task**: Identify relationships between pairs of features in the datasets.<br>

Task 1: Salary vs Age in Employee Dataset<br>
Task 2: Price vs Quantity Sold in Product Sales Dataset<br>
Task 3: Math Score vs Science Score in Student Grades Dataset

Instructions: <br>

Use scatter plots or correlation coefficients to analyze the relationships.<br>
Describe any insights or patterns observed.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 1: Salary vs Age in Employee Dataset
employee_data = pd.DataFrame({
    'Age': [28, 34, 25, 41],
    'Salary': [55000.0, 75000.0, 48000.0, 67000.0]
})
plt.figure(figsize=(5, 4))
sns.scatterplot(x='Age', y='Salary', data=employee_data)
plt.title('Salary vs Age (Employee Dataset)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
corr1 = employee_data['Age'].corr(employee_data['Salary'])
print(f"Correlation (Salary vs Age): {corr1:.2f}")
# Insight: Positive correlation suggests older employees tend to have higher salaries.

# Task 2: Price vs Quantity Sold in Product Sales Dataset
product_data = pd.DataFrame({
    'Price': [10, 12, 15, 14, 13, 16, 18, 100],
    'Quantity Sold': [200, 180, 160, 170, 175, 150, 140, 50]
})
plt.figure(figsize=(5, 4))
sns.scatterplot(x='Price', y='Quantity Sold', data=product_data)
plt.title('Price vs Quantity Sold (Product Sales Dataset)')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.show()
corr2 = product_data['Price'].corr(product_data['Quantity Sold'])
print(f"Correlation (Price vs Quantity Sold): {corr2:.2f}")
# Insight: Negative correlation suggests higher prices are associated with lower sales.

# Task 3: Math Score vs Science Score in Student Grades Dataset
student_data = pd.DataFrame({
    'Math Score': [55, 60, 65, 70, 75, 80, 85, 90, 100, 95],
    'Science Score': [58, 62, 67, 72, 76, 81, 86, 91, 99, 94]
})
plt.figure(figsize=(5, 4))
sns.scatterplot(x='Math Score', y='Science Score', data=student_data)
plt.title('Math Score vs Science Score (Student Grades Dataset)')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.show()
corr3 = student_data['Math Score'].corr(student_data['Science Score'])
print(f"Correlation (Math Score vs Science Score): {corr3:.2f}")
# Insight: Strong positive correlation suggests students who score well in Math also tend to score well in Science.
```
