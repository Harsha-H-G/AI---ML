Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

```python
# Write your code from here

# Create a histogram for the age feature from a dataset

import numpy as np
import matplotlib.pyplot as plt

# Example synthetic age data
age = np.random.normal(loc=35, scale=10, size=200)

plt.figure(figsize=(8, 5))
plt.hist(age, bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.show()

# Interpretation:
# If the histogram is bell-shaped and centered, the age feature is normally distributed.
# If it is skewed left or right, the distribution is not symmetric.
# Peaks or gaps may indicate clusters or missing age groups in the data.
```

Title: Boxplot for a Single Feature <br>

Question 2: Generate a boxplot for the salary feature and identify any outliers.

```python
# Write your code from here
# Generate a boxplot for the salary feature and identify any outliers

import numpy as np
import matplotlib.pyplot as plt

# Example synthetic salary data
np.random.seed(0)
salary = np.append(np.random.normal(loc=50000, scale=8000, size=190), [100000, 120000, 15000])  # add some outliers

plt.figure(figsize=(8, 5))
plt.boxplot(salary, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
plt.title('Boxplot of Salary')
plt.xlabel('Salary')
plt.grid(axis='x', alpha=0.3)
plt.show()

# Outliers are shown as points beyond the whiskers of the boxplot.
# These represent unusually high or low salary values compared to the rest of the data.
```

Title: Scatter Plot to Analyze Relationship<br>

Question 4: Create a scatter plot for the weight and height features to determine if there is a trend.

```python
# Write your code from here
# Create a scatter plot for the weight and height features to determine if there is a trend

import numpy as np
import matplotlib.pyplot as plt

# Example synthetic data for height and weight
np.random.seed(2)
height = np.random.normal(loc=170, scale=8, size=200)
weight = 0.45 * height + np.random.normal(loc=0, scale=5, size=200) + 40  # add some correlation

plt.figure(figsize=(8, 5))
plt.scatter(height, weight, alpha=0.7, color='purple', edgecolor='k')
plt.title('Scatter Plot of Weight vs Height')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(alpha=0.3)
plt.show()

# If the points form an upward trend, there is a positive correlation between height and weight.
# If the points are scattered randomly, there is little or no correlation.
```

Title: Correlation Heatmap<br>

Question 5 : Generate a correlation heatmap for a dataset with multiple features (e.g., height ,weight , age ) and explain the correlations observed.

```python
# Write your code from here
# Generate a correlation heatmap for a dataset with multiple features (e.g., height, weight, age)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example synthetic data
np.random.seed(3)
age = np.random.normal(loc=35, scale=10, size=200)
height = np.random.normal(loc=170, scale=8, size=200)
weight = 0.45 * height + 0.2 * age + np.random.normal(loc=0, scale=5, size=200) + 40

# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'Height': height,
    'Weight': weight
})

# Compute the correlation matrix
corr = df.corr()

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Age, Height, Weight')
plt.show()

# Explanation:
# Values close to 1 or -1 indicate strong positive or negative correlation.
# For example, if Weight and Height have a high positive value, they increase together.
# Low or near-zero values indicate little to no linear relationship.
```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python
# Write your code from here
# Calculate the mean, median, standard deviation, skewness, and kurtosis for the temperature feature and discuss the results

import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# Example synthetic temperature data
np.random.seed(4)
temperature = np.random.normal(loc=22, scale=3, size=200)

# Calculate statistics
mean_temp = np.mean(temperature)
median_temp = np.median(temperature)
std_temp = np.std(temperature)
skew_temp = skew(temperature)
kurt_temp = kurtosis(temperature)

print(f"Mean: {mean_temp:.2f}")
print(f"Median: {median_temp:.2f}")
print(f"Standard Deviation: {std_temp:.2f}")
print(f"Skewness: {skew_temp:.2f}")
print(f"Kurtosis: {kurt_temp:.2f}")

# Discussion:
# - Mean and median close together indicate a symmetric distribution.
# - Standard deviation shows the spread of temperature values.
# - Skewness near 0 means the distribution is roughly symmetric.
# - Kurtosis near 0 means the distribution is close to normal (mesokurtic).
```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
# Write your code from here
# Compute the covariance and correlation between price and demand in a dataset.
# Explain what these metrics indicate.

import numpy as np
import pandas as pd

# Example synthetic data for price and demand
np.random.seed(5)
price = np.random.normal(loc=50, scale=10, size=200)
# Simulate demand inversely related to price, with some noise
demand = 300 - 4 * price + np.random.normal(loc=0, scale=15, size=200)

# Create a DataFrame
df = pd.DataFrame({
    'Price': price,
    'Demand': demand
})

# Compute covariance
covariance = np.cov(df['Price'], df['Demand'])[0, 1]

# Compute correlation
correlation = np.corrcoef(df['Price'], df['Demand'])[0, 1]

print(f"Covariance between price and demand: {covariance:.2f}")
print(f"Correlation between price and demand: {correlation:.2f}")

# Explanation:
# - Covariance indicates the direction of the linear relationship (negative means as price increases, demand decreases).
# - Correlation (ranges from -1 to 1) shows both direction and strength; values near -1 or 1 indicate strong relationships.
# - Here, a strong negative correlation means higher prices are associated with lower demand.
```

Title: Pair Plot for Multivariate Analysis<br>

Question 8: Utilize a pair plot on a dataset to explore the relationships and distributions between
height , weight , and age . What insights can you glean from this visualization?

```python
# Write your code from here
# Utilize a pair plot on a dataset to explore the relationships and distributions between height, weight, and age

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example synthetic data
np.random.seed(10)
age = np.random.normal(loc=35, scale=10, size=200)
height = np.random.normal(loc=170, scale=8, size=200)
weight = 0.45 * height + 0.2 * age + np.random.normal(loc=0, scale=5, size=200) + 40

# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'Height': height,
    'Weight': weight
})

# Create the pair plot
sns.pairplot(df, diag_kind='kde')
plt.suptitle('Pair Plot: Age, Height, Weight', y=1.02)
plt.show()

# Insights:
# - Diagonal plots show the distribution of each feature.
# - Off-diagonal plots show pairwise relationships.
# - If scatter plots show an upward trend, features are positively correlated.
# - If there is no clear pattern, features are weakly or not correlated.
```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python
# Write your code from here
# Apply PCA on a dataset with multiple features (e.g., x1, x2, x3, x4) and reduce it to two principal components. Visualize the data in the new feature space.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Example synthetic data with 4 features
np.random.seed(20)
x1 = np.random.normal(0, 1, 200)
x2 = 2 * x1 + np.random.normal(0, 1, 200)
x3 = np.random.normal(5, 2, 200)
x4 = 0.5 * x3 + np.random.normal(0, 1, 200)

df = pd.DataFrame({
    'x1': x1,
    'x2': x2,
    'x3': x3,
    'x4': x4
})

# Apply PCA to reduce to 2 principal components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df)

# Create a DataFrame for the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Visualize the data in the new feature space
plt.figure(figsize=(8, 6))
plt.scatter(pc_df['PC1'], pc_df['PC2'], alpha=0.7, c='teal', edgecolor='k')
plt.title('PCA: Projection onto First Two Principal Components')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(alpha=0.3)
plt.show()

# The plot shows how the original data is represented in the space of the first two principal components.
```

Title: Advanced Pair Plot with Hue Parameter<br>

Question 10 : Create a pair plot for height , weight , and age with an added categorical variable gender as the hue to observe different group trends.

```python
# Write your code from here
# Create a pair plot for height, weight, and age with an added categorical variable gender as the hue to observe different group trends

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example synthetic data
np.random.seed(42)
age = np.random.normal(loc=35, scale=10, size=200)
height = np.random.normal(loc=170, scale=8, size=200)
weight = 0.45 * height + 0.2 * age + np.random.normal(loc=0, scale=5, size=200) + 40

# Randomly assign gender
gender = np.random.choice(['Male', 'Female'], size=200)

# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'Height': height,
    'Weight': weight,
    'Gender': gender
})

# Create the pair plot with hue
sns.pairplot(df, vars=['Age', 'Height', 'Weight'], hue='Gender', diag_kind='kde', palette='Set2')
plt.suptitle('Pair Plot: Age, Height, Weight by Gender', y=1.02)
plt.show()

# This visualization helps observe if trends or distributions differ between genders.
```
