**Univariate Analysis (Single Feature Analysis)**<br>

Objective: Understand the distribution and characteristics of a single feature using histograms boxplots, and violin plots.<br>

Dataset: Use the Iris dataset or any dataset with multiple numerical features.


Title: Histogram<br>

Task 1: Plot a histogram of the petal length feature.

```python
# Write your code from here
# Univariate Analysis using the Iris dataset

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Display first few rows to verify
print(df.head())
```

Title: Boxplots<br>

Task 2: Plot a boxplot of the petal length feature.

```python
# Write your code from here
# Plot a boxplot of the petal length feature

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['petal length (cm)'], color='lightblue')
plt.title('Boxplot of Petal Length (cm)')
plt.xlabel('Petal Length (cm)')
plt.show()
```

Title: Violin Plots<br>

Task 3: Plot a violin plot of the petal length feature.

```python
# Write your code from here
# Plot a violin plot of the petal length feature

plt.figure(figsize=(6, 4))
sns.violinplot(x=df['petal length (cm)'], color='lightgreen')
plt.title('Violin Plot of Petal Length (cm)')
plt.xlabel('Petal Length (cm)')
plt.show()
```

**Bivariate Analysis (Relationships Between Features)**<br>

Objective: Explore relationships between two features using scatter plots and correlation heatmaps.

Title: Scatter Plots<br>

Task 1: Create a scatter plot between sepal length and sepal width.<br>
Task 2: Scatter plot between petal length and petal width.<br>
Task 3: Scatter plot between sepal length and petal length.<br>

```python
# Write your code from here
# Create scatter plots for the specified feature pairs in the Iris dataset

plt.figure(figsize=(16, 4))

# Task 1: Sepal length vs Sepal width
plt.subplot(1, 3, 1)
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['target'], palette='Set1', legend=False)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Task 2: Petal length vs Petal width
plt.subplot(1, 3, 2)
sns.scatterplot(x=df['petal length (cm)'], y=df['petal width (cm)'], hue=df['target'], palette='Set1', legend=False)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')

# Task 3: Sepal length vs Petal length
plt.subplot(1, 3, 3)
sns.scatterplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue=df['target'], palette='Set1')
plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')

plt.tight_layout()
plt.show()
```

Title: Correlation Heatmaps<br>

Task 1: Generate a correlation heatmap of the dataset.<br>
Task 2: Highlight correlation between sepal length and petal length.<br>
Task 3: Highlight correlation between petal width and petal length.<br>

```python
# Write your code from here
# Generate a correlation heatmap of the Iris dataset and highlight specific correlations

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Compute correlation matrix for numerical features
corr = df.iloc[:, :-1].corr()  # Exclude the target column

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Highlight correlation between sepal length and petal length
sepal_petal_corr = corr.loc['sepal length (cm)', 'petal length (cm)']
plt.text(0.5, 0, f"{sepal_petal_corr:.2f}", ha='center', va='center', color='red', fontsize=14, fontweight='bold')

# Highlight correlation between petal width and petal length
petal_corr = corr.loc['petal width (cm)', 'petal length (cm)']
plt.text(2, 3, f"{petal_corr:.2f}", ha='center', va='center', color='red', fontsize=14, fontweight='bold')

plt.title('Correlation Heatmap of Iris Dataset')
plt.show()

print(f"Correlation between sepal length and petal length: {sepal_petal_corr:.2f}")
print(f"Correlation between petal width and petal length: {petal_corr:.2f}")
```

**Multivariate Analysis (Higher-Dimensional Data Relationships)**<br>
Objective: Analyze relationships in higher-dimensional data using pair plots and PCA.

Title: Pair Plots<br>

Task 1: Create a pair plot for the Iris dataset.<br>
Task 2: Focus on a subset of features (e.g., only petal dimensions).<br>
Task 3: Exclude one class to observe differences.

```python
# Write your code from here
# Create a pair plot for the Iris dataset
import seaborn as sns
import matplotlib.pyplot as plt

# Task 1: Pair plot for all features
sns.pairplot(df, hue='target', diag_kind='kde', palette='Set1')
plt.suptitle('Pair Plot of All Iris Features', y=1.02)
plt.show()

# Task 2: Pair plot for only petal dimensions
sns.pairplot(df, vars=['petal length (cm)', 'petal width (cm)'], hue='target', diag_kind='kde', palette='Set2')
plt.suptitle('Pair Plot: Petal Length vs Petal Width', y=1.02)
plt.show()

# Task 3: Exclude one class (e.g., target != 0)
df_excl = df[df['target'] != 0]
sns.pairplot(df_excl, vars=['petal length (cm)', 'petal width (cm)'], hue='target', diag_kind='kde', palette='Set2')
plt.suptitle('Pair Plot (Excluding Class 0): Petal Length vs Petal Width', y=1.02)
plt.show()
```

Title: Principal Component Analysis (PCA)<br>

Task 1: Perform PCA and plot the first two principal components.<br>
Task 2: Visualize explained variance by each component.<br>
Task 3: Retain more components and visualize in 3D.

```python
# Write your code from here
# Perform PCA and plot the first two principal components
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401

# Prepare features (exclude target)
features = df.drop(columns=['target'])

# Task 1: Perform PCA and plot the first two principal components
pca = PCA(n_components=2)
components_2d = pca.fit_transform(features)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(components_2d[:, 0], components_2d[:, 1], c=df['target'], cmap='Set1', edgecolor='k', alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Principal Components')
plt.colorbar(scatter, ticks=[0, 1, 2], label='Target')
plt.grid(alpha=0.3)
plt.show()

# Task 2: Visualize explained variance by each component
plt.figure(figsize=(6, 4))
plt.bar(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, color='teal')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Principal Components (2D)')
plt.show()

# Task 3: Retain more components and visualize in 3D
pca_3d = PCA(n_components=3)
components_3d = pca_3d.fit_transform(features)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter3d = ax.scatter(components_3d[:, 0], components_3d[:, 1], components_3d[:, 2], c=df['target'], cmap='Set1', edgecolor='k', alpha=0.7)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('PCA: First Three Principal Components')
fig.colorbar(scatter3d, ticks=[0, 1, 2], label='Target')
plt.show()
```

**Statistical Analysis in EDA**<br>

Objective: Calculate basic statistical metrics and explore the relationship between features using correlation and covariance.<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
# Write your code from here
# Calculate descriptive statistics for the Iris dataset

import numpy as np
from scipy.stats import skew, kurtosis

# Task 1: Mean, median, and standard deviation of petal length
petal_length = df['petal length (cm)']
mean_petal_length = np.mean(petal_length)
median_petal_length = np.median(petal_length)
std_petal_length = np.std(petal_length)

print(f"Petal Length - Mean: {mean_petal_length:.2f}, Median: {median_petal_length:.2f}, Std Dev: {std_petal_length:.2f}")

# Task 2: Skewness and kurtosis of sepal width
sepal_width = df['sepal width (cm)']
skew_sepal_width = skew(sepal_width)
kurt_sepal_width = kurtosis(sepal_width)

print(f"Sepal Width - Skewness: {skew_sepal_width:.2f}, Kurtosis: {kurt_sepal_width:.2f}")

# Task 3: Mean, median, and standard deviation of sepal length
sepal_length = df['sepal length (cm)']
mean_sepal_length = np.mean(sepal_length)
median_sepal_length = np.median(sepal_length)
std_sepal_length = np.std(sepal_length)

print(f"Sepal Length - Mean: {mean_sepal_length:.2f}, Median: {median_sepal_length:.2f}, Std Dev: {std_sepal_length:.2f}")
```

Title: Correlation & Covariance<br>

Task 1: Compute correlation between sepal length and petal length.<br>
Task 2: Compute covariance between petal width and sepal width.<br>
Task 3: Determine the most correlated pair of features.<br>

```python
# Write your code from here
# Compute correlation & covariance for the Iris dataset

# Task 1: Correlation between sepal length and petal length
corr_sepal_petal = df['sepal length (cm)'].corr(df['petal length (cm)'])
print(f"Correlation between sepal length and petal length: {corr_sepal_petal:.2f}")

# Task 2: Covariance between petal width and sepal width
cov_petal_sepal = df['petal width (cm)'].cov(df['sepal width (cm)'])
print(f"Covariance between petal width and sepal width: {cov_petal_sepal:.2f}")

# Task 3: Most correlated pair of features
corr_matrix = df.iloc[:, :-1].corr().abs()  # Exclude 'target', use absolute values
np.fill_diagonal(corr_matrix.values, 0)    # Ignore self-correlation
max_corr = corr_matrix.max().max()
most_corr_pair = corr_matrix.stack().idxmax()
print(f"Most correlated pair: {most_corr_pair} with correlation {max_corr:.2f}")
```
