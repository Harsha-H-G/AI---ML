Polynomial Regression

Task 1: Apply polynomial regression to a dataset showing the relationship between age and income.

```python
# Apply polynomial regression to a dataset showing the relationship between age and income

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Example dataset: Age vs Income (Income in $1000s)
np.random.seed(42)
age = np.linspace(18, 65, 30)
income = 15 + 2.5 * age - 0.04 * age**2 + np.random.normal(0, 5, age.shape)
df = pd.DataFrame({'Age': age, 'Income': income})

# Prepare features
X = df[['Age']]
y = df['Income']

# Polynomial transformation (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

# Plot results
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Polynomial Fit (Degree 2)')
plt.xlabel('Age')
plt.ylabel('Income ($1000s)')
plt.title('Polynomial Regression: Age vs Income')
plt.legend()
plt.show()
```

Task 2: Explore and visualize a higher-degree polynomial fitting compared to a simple linear regression.

```python
# Explore and visualize a higher-degree polynomial fitting compared to a simple linear regression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Use the same dataset from Task 1
X = df[['Age']]
y = df['Income']

# Simple Linear Regression (degree 1)
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_linear = lin_reg.predict(X)

# Polynomial Regression (degree 4)
poly_high = PolynomialFeatures(degree=4)
X_poly_high = poly_high.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly_high, y)
y_pred_poly = poly_reg.predict(X_poly_high)

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred_linear, color='green', linewidth=2, label='Linear Fit (Degree 1)')
plt.plot(X, y_pred_poly, color='red', linewidth=2, label='Polynomial Fit (Degree 4)')
plt.xlabel('Age')
plt.ylabel('Income ($1000s)')
plt.title('Linear vs Higher-Degree Polynomial Regression: Age vs Income')
plt.legend()
plt.show()
```
